```python
import bs4
import urllib
```

```python
url = "http://www.fasttrack.co.uk/league-tables/tech-track-100/league-table/"
page = urllib.request.urlopen(url)
soup = bs4.BeautifulSoup(page, 'lxml')
```

```python
table_class = soup.find('table', {'class': 'tableSorter'})
results = table_class.findAll('tr')
results
```

```python
# Creating a list to append the results

rows = []
rows.append(['Rank', 'Company Name', 'Web Page', 'Description', 'Location', 'Year End', 'Annual Sales Rise over 3 years', 'Latest Sales', 'Staff', 'Comments'])
rows
```

```python
def get_web_page(comp_url):
    url_comp = urllib.request.urlopen(comp_url)
    soup_comp = bs4.BeautifulSoup(url_comp, 'lxml')
    table = soup_comp.find('table').findAll('tr')[-1]
    website = table.find('a').get('href')
    return website

for result in results:
    data = result.findAll('td')
    if len(data) == 0:
        continue
        
    rank = data[0].text.strip()
    company_name = data[1].find('span').text.strip()
    web_page = get_web_page(data[1].find('a').get('href'))
    description = data[1].getText().replace(company_name, '')
    location = data[2].text.strip()
    year_end = data[3].text.strip()
    annual_rise = data[4].text.strip()
    latest_sales = data[5].text.strip('*').replace(',', '')
    staff = data[6].text.strip()
    comments = data[7].text.strip()
    
    rows.append([rank, company_name, web_page, description, location, year_end, annual_rise, latest_sales, staff, comments])
```

```python
# Writing to an output file

import csv

with open('tech-track-100.csv', 'w') as f_output:
    csv_output = csv.writer(f_output)
    csv_output.writerows(rows)
```
